```python
# Import libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib

plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline 

matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

df = pd.read_csv(r'C:\Users\MAXNET\Desktop\Portfoilo Project\archive\movies.csv')
```

```python
# Show the data

df.head()
```

```python
# Check if there is missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Data types for our columns

print(df.dtypes)
```

```python
# Changing data type for (budget & gross)

df['budget'] = df['budget'].fillna(0).astype('int64')
df['gross'] = df['gross'].fillna(0).astype('int64')
```

```python
df
```

```python
# Creating a correct year column

df['yearcorrect'] = df['released'].str.extract(r'(\d{4})')
df
```

```python
# Sort the values by Gross

df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
pd.set_option('display.max_rows', None)
```

```python
pd.reset_option('display.max_rows')
```

```python
# Drop Duplicates

df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
# Scatter Plot With Budget vs Gross

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earning')
plt.xlabel('Budget')
plt.ylabel('Gross Earning')
plt.show()
```

```python
# Plot The Budget vs Gross Using Seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "red"}, line_kws={"color": "blue"})
plt.show()
```

```python
#Compute the Pearson correlation matrix using only numeric columns

df.corr(method='pearson', numeric_only=True)
```

```python
# Calculate the Pearson correlation matrix for numeric columns only 
# and visualize it as a heatmap with annotated correlation values

correlation_matrix = df.corr(method='pearson', numeric_only=True)

sns.heatmap(correlation_matrix, annot = True)

plt.title('Correlation Matrix For Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
# Make a copy of the original DataFrame and turn text columns into numeric codes for analysis

df_numerized = df.copy()

for col_name in df_numerized.columns:
    if df_numerized[col_name].dtype == 'object':
        df_numerized[col_name] = df_numerized[col_name].astype('category').cat.codes  

df_numerized
```

```python
# Create and display a heatmap of the Pearson correlation matrix 
# to show relationships between numeric movie features

correlation_matrix = df_numerized.corr(method='pearson', numeric_only=True)

sns.heatmap(correlation_matrix, annot = True)

plt.title('Correlation Matrix For Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
# Quick view of the correlation matrix for all numeric columns

df_numerized.corr()
```

```python
# Get the full correlation matrix

correlation_mat = df_numerized.corr()

# Flatten the matrix into pairs of feature correlations

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
# Sort correlation pairs by their values

sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
# Select and display only the feature pairs with strong positive correlation (> 0.5)

high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
